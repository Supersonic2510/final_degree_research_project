```python
import numpy as np
import struct
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import json
from visual_utils import visualize_lidar_data_3D_static, visualize_lidar_data_3D
```

```python
def load_lidar_data(bin_file):
    with open(bin_file, 'rb') as f:
        data = f.read()
    points = np.frombuffer(data, dtype=np.float32)
    points = points.reshape(-1, 4)  # Each point is (x, y, z, r)
    return points
```

```python
def plot_statistics(data, column_names):
    # Convert to pandas DataFrame
    df = pd.DataFrame(data, columns=column_names)
    
    # Compute statistics
    print(df.describe())
```

```python
# Example usage:
bin_file = 'runs/run_20240730-135117/lidar.bin'  # Replace with your actual file path
predict_json = 'runs/run_20240730-135117/predictions_0.json'  # Replace with 
with open(predict_json, 'r') as file:
    predict_dict = json.load(file)
points = load_lidar_data(bin_file)
```

```python
visualize_lidar_data_3D_static(points, predict_dict['ref_boxes'], predict_dict['ref_labels'], predict_dict['ref_scores'])
```

```python
visualize_lidar_data_3D(points, predict_dict['ref_boxes'], predict_dict['ref_labels'], predict_dict['ref_scores'], window=True)
```

```python
plot_statistics(points, ['X', 'Y', 'Z', 'R'])
```
